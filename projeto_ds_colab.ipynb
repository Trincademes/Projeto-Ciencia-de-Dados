
# üìä Projeto Final: An√°lise de Dados com Python
**Tema:** An√°lise de Desempenho de Alunos com Dados Simulados (`seaborn.tips`)

Este projeto tem como objetivo aplicar t√©cnicas estat√≠sticas com base em dados simulados do ENEM, utilizando a linguagem Python e bibliotecas como Pandas, Seaborn, Matplotlib, Scipy e Numpy.


```python
# Importa√ß√£o das bibliotecas necess√°rias
import pandas as pd                # Para manipula√ß√£o de dados em formato tabular (DataFrame)
import seaborn as sns              # Para visualiza√ß√µes estat√≠sticas elegantes
import matplotlib.pyplot as plt    # Para cria√ß√£o de gr√°ficos
import numpy as np                 # Para opera√ß√µes num√©ricas e estat√≠sticas
from scipy import stats            # Para realizar testes estat√≠sticos (ex: normalidade)

# Carregamento do dataset de exemplo da biblioteca Seaborn
df = sns.load_dataset('tips')     # Dataset 'tips' simulado como se fossem dados de alunos

# Visualiza√ß√£o das primeiras linhas do DataFrame
df.head()                         # Mostra as 5 primeiras linhas da tabela
```


## üîç Escalas de Medi√ß√£o
- `total_bill`: Nota final (Escala de raz√£o)
- `sex`: G√™nero do estudante (Escala nominal)
- `day`: Dia da prova (Escala nominal)
- `time`: Per√≠odo da prova (Escala nominal)
- `size`: N√∫mero de disciplinas (Escala de raz√£o)


## üìà Medidas de Tend√™ncia Central

```python

# C√°lculo de medidas de tend√™ncia central
print("M√©dia:", df['total_bill'].mean())     # M√©dia da coluna 'total_bill' (nota final simulada)
print("Mediana:", df['total_bill'].median()) # Mediana das notas
print("Moda:", df['total_bill'].mode()[0])   # Moda das notas
```

## üìä Medidas de Dispers√£o

```python

# C√°lculo de medidas de dispers√£o
print("Desvio padr√£o:", df['total_bill'].std())           # Desvio padr√£o
print("Vari√¢ncia:", df['total_bill'].var())              # Vari√¢ncia
print("Amplitude:", df['total_bill'].max() - df['total_bill'].min())  # Amplitude (m√°ximo - m√≠nimo)
```

## üìâ Visualiza√ß√µes

```python

# Visualiza√ß√µes
sns.histplot(df['total_bill'], kde=True)  # Histograma com linha de densidade
plt.title('Distribui√ß√£o das Notas (total_bill)')
plt.xlabel('Nota')
plt.ylabel('Frequ√™ncia')
plt.show()

sns.boxplot(x='sex', y='total_bill', data=df) # Nome do eixo X e Y
plt.title('Notas por G√™nero')
plt.show()
```

## üß™ Teste de Normalidade (Shapiro-Wilk)

```python

# Teste de normalidade (Shapiro-Wilk)
stat, p = stats.shapiro(df['total_bill'])  # Aplica o teste √† coluna 'total_bill'
print("Estat√≠stica:", stat)
print("p-valor:", p)
if p > 0.05:
    print("Distribui√ß√£o normal")           # Interpreta√ß√£o do resultado
else:
    print("Distribui√ß√£o n√£o normal")
```

## üîó Correla√ß√£o entre Vari√°veis

```python

# Correla√ß√£o entre n√∫mero de disciplinas (size) e nota (total_bill)
correlation = df[['total_bill', 'size']].corr()  # Matriz de correla√ß√£o
print(correlation)

sns.scatterplot(x='size', y='total_bill', data=df)  # Gr√°fico de dispers√£o
plt.title('Correla√ß√£o entre N√∫mero de Disciplinas e Notas')
plt.show()
```

Observamos que as notas n√£o seguem uma distribui√ß√£o normal, e h√° uma correla√ß√£o positiva entre o n√∫mero de disciplinas e a nota final.
